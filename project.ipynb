# GDP

```python
gdp = pd.read_csv('./Website/data/GDP.csv')
colnames = gdp.columns.tolist()
for i, name in enumerate(colnames[4:]):
    name = name[:4]
    colnames[i+4] = name
gdp.columns = colnames
gdp = pd.melt(gdp, id_vars=['Country Name', 'Country Code', 'Series Name', 'Series Code'], var_name='year', value_name='gdp')
gdp['gdp'] = gdp['gdp'].replace('..', np.nan)
gdp['gdp'] = gdp['gdp'].astype(float)
gdp.to_csv('./Website/data/global_gdp.csv', index=False)
gdp
```

```python
gdp_global = gdp[gdp['Country Name'] == 'World']
plot_gdp = px.line(gdp_global, x='year', y='gdp', title='Global GDP')
plot_gdp.write_html('./Website/figures/global_gdp.html')
plot_gdp.show()
```

# Load the country name and region

```python
state = pd.read_csv('./Website/data/all.csv')
state = state[['name','alpha-3','region','sub-region']]
state.head()
```

# Load Life Expectacy

```python
df_LE = pd.read_csv('./Website/data/API_SP.DYN.LE00.IN_DS2_en_csv_v2_4770434.csv',skiprows=4)
df_LE = pd.merge(df_LE,state,left_on='Country Code',right_on='alpha-3',how='left')
df_LE.drop(['2021','Unnamed: 66', 'name', 'alpha-3'],axis=1,inplace=True)
LE_special_country = df_LE[df_LE['region'].isnull()]
LE_normal_country = df_LE[df_LE['region'].notnull()]
df_LE.head()
```

```python
LE_normal_country_melt = pd.melt(LE_normal_country, id_vars=['Country Name','Country Code','Indicator Name','Indicator Code','region','sub-region'], var_name='Year', value_name='Life Expectancy')
LE_special_country_melt = pd.melt(LE_special_country, id_vars=['Country Name','Country Code','Indicator Name','Indicator Code','region','sub-region'], var_name='Year', value_name='Life Expectancy')
LE_normal_country_melt.to_csv('./Website/data/LE_normal_country_melt.csv',index=False)
LE_special_country_melt.to_csv('./Website/data/LE_special_country_melt.csv',index=False)
```

```python
# calculate the average life expectancy of each region
LE_normal_country_average = LE_normal_country_melt.groupby(['region','Year'])['Life Expectancy'].mean().reset_index()
fig = px.line(LE_normal_country_average, x="Year", y="Life Expectancy", color='region',title='Average Life Expectancy of Each Region')
# save fig as html
fig.write_html("./Website/figures/LE_state_average.html")
fig.show()
```

# Fertility

```python
fertility = pd.read_csv("./Website/data/API_SP.DYN.TFRT.IN_DS2_en_csv_v2_4770506.csv",skiprows=4)
fertility.drop(['2021','Unnamed: 66'],axis=1,inplace=True)
fertility = pd.merge(fertility,state,left_on='Country Code',right_on='alpha-3',how='left')
fertility.drop(['name','alpha-3'],axis=1,inplace=True)
fertility_special_country = fertility[fertility['region'].isnull()]
fertility_normal_country = fertility[fertility['region'].notnull()]
fertility.head()
```

```python
# create an animation of fertility rate of each region
fertility_normal_country_melt = pd.melt(fertility_normal_country, id_vars=['Country Name','Country Code','Indicator Name','Indicator Code','region','sub-region'], var_name='Year', value_name='Fertility Rate')
# create an animated map for the fertility rate of each region
fig = px.choropleth(fertility_normal_country_melt, locations="Country Code",
                    color="Fertility Rate", # lifeExp is a column of gapminder
                    hover_name="Country Name", # column to add to hover information
                    animation_frame="Year",
                    color_continuous_scale=px.colors.sequential.Plasma_r,
                    title='Fertility Rate of Each Region',
                    range_color=[min(fertility_normal_country_melt['Fertility Rate']), max(fertility_normal_country_melt['Fertility Rate'])])
fig.write_html("./Website/figures/fertility_state.html")
fig.show()
```
